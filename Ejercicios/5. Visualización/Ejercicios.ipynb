Se adjuntan las series histórica de varios tipos de cambio.

```python
# Tratamiento de datos con pandas
import pandas as pd

# Tratamiento de directorios
from glob import glob
from pathlib import Path

# Tratamiento de fechas
from datetime import datetime
```

Para la extracción de la información, creamos este objeto de dataframes en el que cada posición se llama como cada subyacente.

```python
# --
# Escribir la ruta
# --
ruta = r'../Datos/***.csv'

# --
# Accedemos al directorio
# --
file_path = glob(ruta)
forex = {}

# --
# Creamos el iterable
# --
for file in file_path:
    Asset = Path(file).stem
    Asset = Asset.split("_")[-1]
    forex[Asset] = pd.read_csv(file,sep = ",")
    forex[Asset] = forex[Asset].rename({forex[Asset].columns[1]:'Precio'}, axis = 1)
    
    # Tratamiento de la fecha
    forex[Asset]['Fecha'] = forex[Asset]['Fecha'].apply(lambda x: datetime.strptime(x,"%d/%m/%Y"))
    forex[Asset].set_index('Fecha', inplace=True)
```
