```python
# --
# Importación (importante instalar ambas)
# --
import matplotlib.pyplot as plt
import seaborn as sns
```

## Regresión

```python
import numpy as np

# --
# Definimos los datos
# --
num_points = 20
x = 5 + np.arange(num_points) + np.random.randn(num_points)
y = 10 + np.arange(num_points) + 5 * np.random.randn(num_points)

# --
# Creamos el gráfico de regresión
# --
sns.regplot(x, y)

# --
# Graficamos
# --
plt.show()
```

## Heatmaps

```python
# --
# Definimos los datos
# --
side_length = 10
data = 5 + np.random.randn(side_length, side_length)
data += np.arange(side_length)
data += np.reshape(np.arange(side_length), (side_length, 1))

# --
# Creamos el heatmap
# --
sns.heatmap(data)

# --
# Graficamos
# --
plt.show()
```

# Altair

Interesante para barrer el gráfico y ver el histograma seleccionado.

Más información, en el siguiente [link](https://altair-viz.github.io/gallery/).

```python
# --
# Instalación
# --
!pip install vega_datasets
!pip install altair
```

```python
# --
# Importar librería
# --
import altair as alt
```

```python
# --
# Definimos los datos
# --
from vega_datasets import data
cars = data.cars()


interval = alt.selection_interval()

points = alt.Chart(cars).mark_point().encode(
  x='Horsepower',
  y='Miles_per_Gallon',
  color=alt.condition(interval, 'Origin', alt.value('lightgray'))
).properties(
  selection=interval
)

histogram = alt.Chart(cars).mark_bar().encode(
  x='count()',
  y='Origin',
  color='Origin'
).transform_filter(interval)

points & histogram
```
